```python
import seaborn as sns
import matplotlib.pyplot as plt

# Plot app categories
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Category',color ="Blue")
plt.xticks(rotation=90)
plt.xlabel('App Category')
plt.ylabel('Count')
plt.title('Distribution of App Categories')
plt.show()
```

```python
# Plot ratings
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='Rating', bins=20)
plt.xlabel('Rating')
plt.ylabel('Count')
plt.title('Distribution of Ratings')
plt.show()
```

```python
# Plot reviews
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='Reviews', bins=20)
plt.xlabel('Number of Reviews')
plt.ylabel('Count')
plt.title('Distribution of Reviews')
plt.show()
```

```python
#so there is error of strig and int so first we need to check type of intsalls and rating column
df["Installs"].info()
```

```python
df["Rating"].info()
```

```python
df["Installs"]=df["Installs"].astype(str)
df["Installs"]
```

```python
#as you can see the Installs item have "," and "+" se we replace it with ""
df["Installs"]= df["Installs"].str.replace(",","",regex=True)
df["Installs"]= df["Installs"].str.replace("+","",regex=True)
df["Installs"]
```

```python
df["Installs"] = pd.to_numeric(df["Installs"], errors="coerce")
```

```python
print(df["Installs"],df["Rating"])
```

```python
#corelation
correlation = df['Installs'].corr(df['Rating'])
correlation
```

```python
category_counts = df["Category"].value_counts()
category_counts
print(category_counts.head(5))
```

```python
# Plot app categories
#this bar chart is for dislaying first 5 top categories
category_counts = df["Category"].value_counts().head(5)
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Category',color ="Blue",order=category_counts.index)
plt.xticks(rotation=90)
plt.xlabel('App Category')
plt.ylabel('Count')
plt.title('Distribution of App Categories')
plt.show()
```

```python
#pie chart
plt.figure(figsize=(8, 8))
plt.pie(category_counts, labels=category_counts.index, autopct='%1.1f%%')
plt.title('Top 5 App Categories')
plt.show()
```

```python
#Extract insights from the dataset.
average_ratings=df.groupby("Category")["Rating"].mean()
average_ratings
```

```python
sorted_ratings = average_ratings.sort_values(ascending=False)
top_categories = sorted_ratings.head(5)
top_categories

```

```python
#summarze
# Bar chart of average ratings by category
plt.figure(figsize=(10, 6))
average_ratings.plot(kind='bar')
plt.xlabel('Category')
plt.ylabel('Average Rating')
plt.title('Average Ratings by Category')
plt.xticks(rotation=45)
plt.show()
```

```python
# Scatter plot of ratings vs. downloads
plt.figure(figsize=(8, 6))
plt.scatter(df['Installs'], df['Rating'], alpha=0.5)
plt.xlabel('Installs')
plt.ylabel('Rating')
plt.title('Ratings vs. Installs')
plt.show()
```

```python
# Scatter plot of ratings vs. reviews
plt.figure(figsize=(8, 6))
plt.scatter(df['Reviews'], df['Rating'], alpha=0.5)
plt.xlabel('Reviews')
plt.ylabel('Rating')
plt.title('Ratings vs. Reviews')
plt.show()
```
